```python
def get_pivot_table(df, column_name, columns_to_rename={}):
    if 'UUID' not in columns_to_rename.values():
        columns_to_rename['UUID'] = 'UUID'
    
    # sum GWP and A2GWPtotal (A2) columns
    df['GWP_sum'] = df[['GWP', 'A2GWPtotal (A2)']].sum(axis=1)
    
    df_a1a3 = df[df['Modul'].isin(['A1', 'A2', 'A3'])].copy()
    df_not_a1a3 = df[~df['Modul'].isin(['A1', 'A2', 'A3'])].copy()
    df_a1a3['Modul'] = 'A1A2A3'
    df_a1a3 = df_a1a3.groupby(['UUID', 'Modul']).sum().reset_index()
    df_concat = pd.concat([df_not_a1a3, df_a1a3])
    df_pivot = pd.pivot_table(df_concat, values='GWP_sum', index=['UUID'], columns=[column_name], aggfunc=np.sum, fill_value=0)
    df_pivot = df_pivot.reset_index()
    columns_to_include = {columns_to_rename[k]: k for k in columns_to_rename.keys()}  # swap keys/values in dictionary
    if len(columns_to_include) > 0:
        df_additional = df.rename(columns=columns_to_include)[columns_to_include.values()].groupby('UUID').first().reset_index()
        df_pivot = df_pivot.merge(df_additional, on='UUID', how='left')
        df_pivot = df_pivot.rename(columns={v: k for k, v in columns_to_rename.items()})  # swap keys/values back in dictionary
    return df_pivot
```

```python

def float_to_decimal(num):
    return Decimal(str(num))



```

```python
def pandas_to_dynamodb(df, table_name):
    df = df.fillna(0)
    # convert any floats to decimals
    for i in df.columns:
        print(i)
        print(df[i].dtype)
        datatype = df[i].dtype
        if 'defaultAmount' in df.columns:
            df['defaultAmount'] = df['defaultAmount'].apply(float_to_decimal)
        if datatype == 'float64':
            df[i] = df[i].apply(float_to_decimal)
            print(df[i]);
    # write to dynamodb using batch write operations
    dynamodb = boto3.resource('dynamodb')
    table = dynamodb.Table(table_name)
    with table.batch_writer() as batch:
        for index, row in df.iterrows():
            item = row.to_dict()
            print(row)
            print(item)
            print("Data types of item:")
            for key, value in item.items():
                print(f"{key}: {type(value)}")
            batch.put_item(Item=item)
```
